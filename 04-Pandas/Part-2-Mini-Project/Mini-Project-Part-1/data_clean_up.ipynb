```python
# Dependencies
import pandas as pd
import numpy as np
import chardet
```

```python
# load CSV
file="Resources/2016-FCC-New-Coders-Survey-Data.csv"
chardet.detect(open(file,'rb').read())
```

```python
# Read with pandas
new_coders_pd=pd.read_csv(file,encoding='utf-8', low_memory=False)
new_coders_pd.head()
```

```python
# Inspect all columns
new_coders_pd.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)
reduced_coders_pd=new_coders_pd.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
reduced_coders_pd.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function
reduced_coders_pd=reduced_coders_pd.replace({0.0:"No",1.0:"Yes"})
reduced_coders_pd.head(20)
```

```python
# Calculate total number of respondents in survey
# consider using the len() function
num_of_respondents=len(reduced_coders_pd)
num_of_respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
attended_bootcamp_df=reduced_coders_pd.loc[reduced_coders_pd["AttendedBootcamp"]=="Yes"]
attended_bootcamp_df.head()
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it
average_age=attended_bootcamp_df["Age"].mean()
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
number_of_attendees=attended_bootcamp_df["AttendedBootcamp"].count()
number_of_attendees
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column
has_degree=attended_bootcamp_df["SchoolDegree"].value_counts()
has_degree
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column
yes_has_degree_df=attended_bootcamp_df.loc[(attended_bootcamp_df["SchoolDegree"]=="bachelor's degree")|
                                         (attended_bootcamp_df["SchoolDegree"]=="master's degree (non-professional)")|
                                         (attended_bootcamp_df["SchoolDegree"]=="professional degree (MBA, MD, JD, etc.)")|
                                         (attended_bootcamp_df["SchoolDegree"]=="associate's degree")|
                                         (attended_bootcamp_df["SchoolDegree"]=="Ph.D."),:]

num_degree_holders=yes_has_degree_df["SchoolDegree"].count()
                                         
print(f'The total number of degree holders is {num_degree_holders}')
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
total_gender=attended_bootcamp_df["Gender"].count()
male=attended_bootcamp_df["Gender"].value_counts()['male']
print(f"males={male}")
female=attended_bootcamp_df["Gender"].value_counts()['female']
print(f"females={female}")
non_binary_gender=total_gender-male-female
print(f"non binary gender={non_binary_gender}")

```

```python
# Calculate percentage of respondents who attended a bootcamp
percent_attended=number_of_attendees/num_of_respondents *100
percent_attended
```

```python
# Calculate percentage of respondents belonging to each gender
percent_male=(male/total_gender)*100
print(f" % male: {percent_male}")

percent_female=(female/total_gender)*100
print(f" % female: {percent_female}")
percent_non_binary=(non_binary_gender/total_gender)*100
print(f" % non binary: {percent_non_binary}")
```

```python
# Calculate percentage of attendees with a college degree
percentage_college_degree=(num_degree_holders/number_of_attendees)*100
percentage_college_degree
```

```python
# Calculate average post-bootcamp salary
average_salary=attended_bootcamp_df["BootcampPostSalary"].mean()
average_salary
```

```python
# Create a new table consolidating above calculations
summary_df=pd.DataFrame({"Total Surveyed":[num_of_respondents],
                         "Total Bootcamp attendees":[number_of_attendees],
                         "% attended bootcamp":[percent_attended],
                         "Average Age":[average_age],
                         "Has a degree":[percentage_college_degree],
                         "% Male":[percent_male],
                         "% Female":[percent_female],
                         "% Non Gender Specific":[percent_non_binary],
                         "Average Post Bootcamp Salary":[average_salary]
                        })

summary_df
```

```python
# Improve formatting before outputting spreadsheet

summary_df["% attended bootcamp"] = summary_df["% attended bootcamp"].map("{0:,.2f}%".format)
summary_df["Average Age"] = summary_df["Average Age"].map("{0:,.2f}".format)
summary_df["Has a degree"] = summary_df["Has a degree"].map("{0:,.2f}%".format)
summary_df["% Male"] = summary_df["% Male"].map("{0:,.2f}%".format)
summary_df["% Female"] = summary_df["% Female"].map("{0:,.2f}%".format)
summary_df[ "% Non Gender Specific"] = summary_df[ "% Non Gender Specific"].map("{0:,.2f}%".format)
summary_df["Average Post Bootcamp Salary"] = summary_df["Average Post Bootcamp Salary"].map("${0:,.0f}".format)


summary_df




```

```python
# Export to Excel
# Use df.to_excel to export to excel. Don't include the indexes
summary_df.to_excel("output/CleanUp_Part1_Output_clemi.xlsx",index=False)

```
